```python
# Monkey Banana Problem using DFS

# State: (Monkey_Pos, Box_Pos, Monkey_Status, Has_Banana)

def is_goal(state):
    return state[3] == True

def get_neighbors(state):
    monkey, box, status, banana = state
    neighbors = []

    # Move monkey
    for pos in ['A', 'B', 'C']:
        if pos != monkey and status == 'on_floor':
            neighbors.append((pos, box, 'on_floor', banana))

    # Push box
    if monkey == box and status == 'on_floor':
        for pos in ['A', 'B', 'C']:
            if pos != box:
                neighbors.append((pos, pos, 'on_floor', banana))

    # Climb box
    if monkey == box and status == 'on_floor':
        neighbors.append((monkey, box, 'on_box', banana))

    # Grasp banana
    if monkey == 'C' and box == 'C' and status == 'on_box':
        neighbors.append((monkey, box, 'on_box', True))

    return neighbors

def dfs(state, visited):
    if is_goal(state):
        return [state]

    visited.add(state)

    for neighbor in get_neighbors(state):
        if neighbor not in visited:
            path = dfs(neighbor, visited)
            if path:
                return [state] + path

    return None

# Initial State
initial_state = ('A', 'B', 'on_floor', False)

solution = dfs(initial_state, set())

if solution:
    print("Solution Found!\n")
    for step in solution:
        print(step)
else:
    print("No Solution Found")
```
