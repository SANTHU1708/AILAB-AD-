```python
import heapq

def uniform_cost_search(graph, start, goal):
    # Priority queue stores (cumulative_cost, node, path) and always pops the lowest cost
    priority_queue = [(0, start, [start])]
    visited = set()

    while priority_queue:
        cost, node, path = heapq.heappop(priority_queue)

        if node == goal:
            return cost, path

        if node in visited:
            continue

        visited.add(node)

        for neighbor, edge_cost in graph.get(node, []):
            if neighbor not in visited:
                # Add neighbor to queue with updated total cost and current path
                heapq.heappush(priority_queue, (cost + edge_cost, neighbor, path + [neighbor]))

    return float("inf"), []

graph = {
    'A': [('B', 1), ('C', 4)],
    'B': [('C', 2), ('D', 5)],
    'C': [('D', 1)],
    'D': []
}

cost, path = uniform_cost_search(graph, 'A', 'D')

print("Minimum cost:", cost)
print("Path:", path)
```
