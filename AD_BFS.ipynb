```python
from collections import deque

# Graph representation (same tree)
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': [],
    'D': ['I', 'J'],
    'E': ['G', 'H'],
    'F': [],
    'G': [],
    'H': [],
    'I': ['K', 'L'],
    'J': [],
    'K': [],
    'L': []
}

def goal_based_bfs(graph, start, goal):
    queue = deque([(start, [start])])
    visited = []

    while queue:
        node, path = queue.popleft()

        if node not in visited:
            visited.append(node)

            if node == goal:
                return visited, path

            for child in graph[node]:
                queue.append((child, path + [child]))

    return visited, None


# -------- OUTPUT --------
source_node = 'A'
goal_node = 'L'

visited_nodes, path = goal_based_bfs(graph, source_node, goal_node)

vertices = list(graph.keys())
edges = [(parent, child) for parent in graph for child in graph[parent]]
unvisited_nodes = list(set(vertices) - set(visited_nodes))

print("Source Node:", source_node)
print("\nVertices:")
print(vertices)

print("\nEdges:")
print(edges)

print("\nVisited Nodes:")
print(visited_nodes)

print("\nUnvisited Nodes:")
print(unvisited_nodes)

print("\nPath of A:")
print(path)
```
